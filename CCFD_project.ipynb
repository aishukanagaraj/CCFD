```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.externals import joblib

from scipy.io import arff
import numpy as np

#load dataset
data = arff.loadarff('bank.arff')
bank = pd.DataFrame(data[0])

#remove the unwanted columns in the dataset which can affect the result 
#columns don't want to include in the model are deleted

del bank['mcc_cat']
del bank['mcc_ant_cat']
del bank['cep_cat']
del bank['cep_ant_cat']
del bank['pos_entry']
del bank['lim_cred_cat']
del bank['bandeira_cat']
del bank['tp_pessoa_cat']
del bank['velocidade_cat']
bank.head(10)
```

```python
#encoding - ONE HOT ENCODER

bank.flag_fraude_cat[bank.flag_fraude_cat == 'N'] = 0
bank.flag_fraude_cat[bank.flag_fraude_cat == 'S'] = 1
features_bank = pd.get_dummies(bank, columns=['score_cat'])
features_bank.head(10)
```

```python
bank.info()
```

```python
from sklearn.model_selection import train_test_split

X=bank.drop('flag_fraude_cat',axis=1)
y=bank['flag_fraude_cat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Fit regression model
model = ensemble.GradientBoostingRegressor()
model.fit(X_train, y_train)

# Save the trained model to a file so we can use it in other programs
joblib.dump(model, 'model/trained_classifier_model.pkl')

# Find the error rate on the training set
mae = mean_absolute_error(y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.10f" % mae)

# Find the error rate on the test set
mae = mean_absolute_error(y_test, model.predict(X_test))
print("Test Set Mean Absolute Error: %.10f" % mae)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

#reg = linear_model.LinearRegression()
#reg.fit (X_train,y_train)
pred=model.predict(X_test)
mae = mean_absolute_error(y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.10f" % mae)
mae = mean_absolute_error(y_test,pred)
print("Test Set Mean Absolute Error : %.10f" % mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit (X_train,y_train)
pred=model.predict(X_test)
mae = mean_absolute_error(y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.10f" % mae)
mae = mean_absolute_error(y_test,pred)
print("Test Set Mean Absolute Error : %.10f" % mae)
```

```python
from sklearn.linear_model import BayesianRidge
model = BayesianRidge()
model.fit (X_train,y_train)
pred=model.predict(X_test)
mse = mean_absolute_error(y_train, model.predict(X_train))
print("Mean Absolute Error : %.4f" % mse)
mse = mean_absolute_error(y_test,pred)
print("Mean Absolute Error : %.4f" % mse)
```

```python

# Fit regression model
from sklearn.metrics import mean_squared_error
model = ensemble.GradientBoostingRegressor()
model.fit(X_train, y_train)

# Save the trained model to a file so we can use it in other programs
joblib.dump(model, 'model/square_classifier_model.pkl')

# Find the error rate on the training set
reg_mse = mean_squared_error(y_train, model.predict(X_train))
print("Training Set Mean Squared Error: %.10f" % reg_mse)

# Find the error rate on the test set
reg_test_mse = mean_squared_error(y_test, model.predict(X_test))
print("Test Set Mean Squared Error: %.10f" % reg_test_mse)
```
